```python
import numpy as np
import matplotlib.pyplot as plt

M = 9

x_nodes = [2.6,3.5,4.9,5.2,4,3.6,5,6.2,6.6]
y_nodes = [1.7,3.1,6.9,9.4,7.2,1.5,4,2,1.6]

t = np.cos(np.array([(2*k+1)*np.pi/(2*M) for k in range(M)]))
n = np.linspace(-1,1,500)

def lagrange_poly(t,x_nodes,y_nodes):
  p = 0
  N = len(x_nodes)
  for j in range(N):
    lj = 1
    for i in range(N):
      if x_nodes[i]!=x_nodes[j]:
        lj *= (t-x_nodes[i])/(x_nodes[j]-x_nodes[i])
    p += lj*y_nodes[j]

  return p

px = lagrange_poly(n,t,x_nodes)
py = lagrange_poly(n,t,y_nodes)

plt.figure(figsize=(10,10))
plt.plot(x_nodes,y_nodes,'rd')
plt.plot(px,py)
```
